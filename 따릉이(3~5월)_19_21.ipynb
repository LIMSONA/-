### 현재 상태 확인하기

```python
# 번호 +  대여소명이 아닌 경우가 있음을 확인함 ==> 번호 없이 대여소명만 있을 수도 있다!

#2019
print(tmp_station_19.대여소명.map(lambda x: x.split(".")[0]).sort_values().unique()[-10:])
#2020
print(tmp_station_20.대여소명.map(lambda x: x.split(".")[0]).sort_values().unique()[-10:])
#2021
print(tmp_station_21.대여소명.map(lambda x: x.split(".")[0]).sort_values().unique()[-10:])
```

```python
# 사용자 정의 함수 - 대여소명 번호찾기
import re

def name_filter(x):
  filter= '^[0-9]+\.'
  p= re.compile(filter)

  result= p.split(x)[1]
  if result[0]==" ":
    result=result[1:]
    return result
  else:
    return result
```

```python
name_filter('1002. 해공공원(천호동)')
```

### 2019

```python
# 정리된 2019
# '상암센터 정비실' '위트콤' '중랑센터'
tmp_station_19.대여소명=\
tmp_station_19.대여소명.map(lambda x: x if (x.split(".")[0]=="상암센터 정비실")\
                        or(x.split(".")[0]=="위트콤")\
                        or(x.split(".")[0]=="중랑센터")\
                        else name_filter(x))
```
